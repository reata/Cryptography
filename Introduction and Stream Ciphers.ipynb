# 一、导论

## 概述
密码学的应用：网络加密（HTTPS，WPA2，Bluetooth），硬盘文件加密，内容加密（DVD，蓝光光盘），用户认证。

网络加密TLS包括两个部分，建立公钥和安全传输。首先双方通过握手协议（handshake protocol）建立连接取得共享公钥，之后通过共享公钥对数据加密再开始传输。

硬盘文件加密可以看成今天的自己向明天的自己发送了一份文件，某种意义上也是一种网络通信。

发送方通过加密算法和公钥将明文加密为密文，而接收方则通过解密算法和公钥将密文解密为明文。加密算法是公开的，不要使用私有加密算法，整个流程中只有公钥未>知。

密钥可以是一次性密钥（Email），也可以是多次密钥。

## 什么是密码学
密码学除了在安全通信之外，还应用于电子签名和匿名通信。电子签名根据内容而每次不同。匿名通信是双向的，通信双发都可以不知道对方是谁。匿名货币如何同时实>现匿名和防止拷贝。

对于一个选举或者拍卖系统，我们希望所有人的输入（投票、投标）是加密的，而最后只披露最终结果。现实世界中的可行方法是引入第三方可信机构作为中介。而密码>学有一个中心理论：任何通过权威第三方实现的系统，都可以不需要该第三方。

密码学的三个主要步骤：1）精确定义威胁模型。2）提出一个应对方案。3）证明对该方案，任何遵从威胁模型的破解，都会演变为求解一个异常复杂的问题。

## 密码学的历史
历史上的一些加密方案（均已告破）：

1. 替换式密码：随机生成字典，仅考虑无大小写的字幕，共计可生成26！，约288种字典，该字典就是密钥，最古老的方法。凯撒加密是字典固定（+3映射）的替换加密>，不存在密钥，严格意义上不算一种加密方案；破解方案：根据英文字母的词频排序寻找对应，最高频的字母e在英文中出现的频率是12.7%，第二t的频率是9.1%，第三a>的频率是8.1%，因此密文中最高频的字母很可能对应的原文是e。之后结合高频字母对，比如he, an, in, th等，反复试错。
2. 维吉尼亚密码：密钥是一个单词，将单词重复多次达到和明文同样长度。每个数位上求（重复密钥对应字母+明文对应字母）% 26；破解方案，首先假设一个密钥长度>，按该长度将密文分为多组，每组对应序号的字母是一个集合，该集合中词频最高的字母对应的原文很可能是e，于是可以计算对应的密钥字母。
3. 旋转盘加密：实质上还是替换加密，对单层旋转盘，每输入一个字幕，旋转盘转动一个单位，字典也随之移动一个单位。二战时期著名的Enigma是多层旋转盘，并且每
层的转动速率不一致。
4. 数字加密：机械加密时代的结束。随着计算能力不断增强，机械加密，包括早期的数字加密都可以暴力破解。

# 二、流密码
## 信息论安全性与一次性密码本
密码的定义，三个输入（密钥集合K，明文集合M，密文集合C），两个高效（在明文多元级别的时间内完成计算）的算法（加密算法E，解密算法D）。一个有效的密码需要
满足对任意在明文集合中的m和任意在密钥集合中的k，D（k，E（k，m））=m。

加密算法常常具有随机性，给定明文和密钥，可产生不同的密文。而解密算法则是唯一性的。

一次性密码本：密钥集合、明文集合和密文集合都是二进制字符串。密钥随机生成，是一个和明文长度相同的二进制字符串。加密算法是对密钥和明文求逻辑异或，而解>密算法是对密钥和密文求逻辑异或。

一次性密码本在实践中几乎无法运用，因为传输明文之前，需要传输一段和明文等长的密钥，这个密钥要如何加密？

一个具有完善保密性的密码，简单来说，密文不能透露除总长度外任何与明文相关的信息。按照香农的信息论安全性，对等长的任意两条明文，选取任意密钥，通过加密>算法得到服从同样分布密文的概率相等。因此拿到这条密文是无法确定来自哪条明文的。

可以证明，一次性密码本具有完善保密性。无法只用密文进行破解。

香农还有另一个证明，如果一个密码具有完善保密性，那么密钥的长度至少要等于明文。

## 流密码与伪随机生成器
一次性密码本的实际实现——流密码：生成一小段随机数序列（seed）作为密钥，然后基于确定性算法的伪随机生成器根据该密钥，生成与明文同长、看上起呈现随机形式>的伪随机数序列，进而与明文做异或运算。

流密码不具备完善保密性，因此我们需要一个关于保密性的新定义。而这个保密性与伪随机生成器密切相关。

保密性包含不可预知性。如果伪随机数生成器生成的一串数位，给出前i位，即可以根据某种算法推测（大于50%的概率）之后的位，则表明该伪随机数生成器可预知，这>是不保密的。

几个可预知的伪随机生成器：1）线性同余方法：N<sub>j+1</sub> = (A * N<sub>j</sub> + B) % M。2）glibc随机。

可忽略值是一个标量。在实际中，大于1/230（1GB数据量中可能发生），则不可忽略。小于1/280（公钥的生命周期内不会发生）则可以忽略。

## 对流密码与一次性密码本的攻击
一次性密码本的密钥多次使用就会不安全，也即是说，不要使用两次性密码本。

WEP加密方法就存在上述问题，一定周期后公钥会重复。另外周期内相邻两个时段的公钥高度相似，监听一定数量的密文就可以破解公钥。

对于硬盘文件加密，不要使用流密码，因为硬盘文件每次的改动都极小。

一次性密码本就有可塑性，用猜测的密钥对密文做异或运算，可以对解密后的明文造成定向影响。从而改变了明文的完整性。而改变密文的这个过程无法监控。

## 现实世界中的流密码
两个已经被破解的伪随机数生成器：1）RC4，针对软件，seed的大小为128位，将该seed拓展为2048位，作为伪随机生成器的内部参数。每次循环生成1字节。2）CSS，针>对硬件，基于线性反馈移位存储器，seed的大小为5字节（40位），DVD、GSM、Bluetooth都是通过这个方式加密的，因此都不安全。

现代使用的流密码eStream：该方法的伪随机数生成器除了需要seed之外，还需要一个nonce（不重复的随机数）作为输入。对于指定key（seed），每个nonce只出现最多>一次，因此key可以重复使用，因为nonce不同，通过伪随机生成的序列就会不同。目前除了穷举还没有找到更好的破解办法，是相对安全的。

Salsa20作为一种eStream，既可以用于软件，也可以用于硬件。Seed的大小为128位或256位，nonce大小为64位，在x86平台上加密速度非常快（硬件指令集SSE2包含了算>法所需要的所有计算）。
